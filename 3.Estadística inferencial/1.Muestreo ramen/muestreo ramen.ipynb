## Muestreo con reposici칩n

```python
styles = df_ramen['Brand'].unique().tolist()
print(f'Existen {len(styles)} estilos de ramen en el dataset')
```

```python
df_ramen['Brand'].sample(n=20, replace=True).tolist()
```

### Realizaremos 500 extracciones con reposici칩n de muestras de tama침o 100

```python
muestras_estilos = []
for i in range(500):
    muestras_estilos.append(df_ramen['Brand'].sample(n=100, replace=True).tolist())

print(f'Cantidad de muestras: {len(muestras_estilos)}')
print(f'Tama침o de cada de muestras: {len(muestras_estilos[0])}')
print(f'Total de datos: {len(muestras_estilos) * len(muestras_estilos[0]):,}')
muestras_estilos
```

### Realizaremos 500 extracciones sin reposici칩n de muestras de tama침o 100

```python
muestras_estilos2 = []
for i in range(500):
    muestras_estilos2.append(df_ramen['Brand'].sample(n=100).tolist())

print(f'Cantidad de muestras: {len(muestras_estilos2)}')
print(f'Tama침o de cada de muestras: {len(muestras_estilos2[0])}')
print(f'Total de datos: {len(muestras_estilos2) * len(muestras_estilos2[0]):,}')
muestras_estilos2
```
